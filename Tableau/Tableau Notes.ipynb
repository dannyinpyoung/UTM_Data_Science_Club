## Creating Calculated Field

Original data may not have the data we want. Use calculated field to arrive at the data we want. 

Say that we want price * quantity and they are separate measures. Go to analysis and then scroll down to create calculated field. Type `[field_name]` `mathematical operator` `[field_name]`. We can rename this new calculated field by double clicking the name. We can edit the fields by clicking the down arrow next to the field and click edit. To delete a field, click field down arrow and click delete. 

## Creating Calculated Field on A Shelf

Create a calculation without going to an external dialogue box. 
We can double click an empty field in columns or rows and then enter the equation we require for the calculated field. If we want it to become a normal calculated field, we can drag it to the measures and rename it then. 

## Adding a Table Calculation

Create a table first of all. 
Change how the data is calculated and displayed by the table. Click the pill holding the values and click "quick table calculation" to change the calculations seen in the table. 

Click "add table calculation" to get a dialogue box to have a more comprehensive table calculation menu. "Down" versus "across" is just columns versus row calcualtions. 

## Level of detail expressions. 

Specifies the level of detail for which we want the calculation to occur. Say for example, we want to see the total sales differenced on the average sales, so now we need to conduct a calculation for two levels. 

To create LOD: 
- Double click an empty pill (usually rows) 
- For example, we would do [Order Total] - AVG[Order Total] 
- This would not work because we did not specify the orders of detail. 
- Therefore we need to do this: `[Order Total] - {AVG([Order Total])}`
- Why doesn't that work? This is because we need to get it right the first time and redo it before making an error. 
- We can drag this pill to the measures to create a new field

## Create Wildcard Filter

Show values that follow a specific pattern using wildcard filters. 

Click on pill scroll down => Filter => Wildcard

Using: 

`Contains`: Will filter for all of the items that have the following filters. 

`Starts With`: Those that will start with the value inputted. 

`Ends With`: Values that end with the value inputted. 

Instead of using exact, just change the selection filter.

## Filter data using parameters

We would right click the pill of choice => click create => parameters. Drag the pill of choice to the filter tab and then go to `tab` and then choose `by field` and choose `by parameter`. 

Go to parameter panel and then right click and show parameter control. It will be a slider to change the number of values we'd like to see. 

First create parameter by right clicking measure => Write down filter values by clicking "allowable values" => and then right click and show parameter control (sets up filter) => Create calculated field => Drag calculated field to measure => Use the code: 

```
CASE [Parameter name]
When 1 then Parameter_filter
When 2 then Parameter_filter
...
END
```

Then drag the calculated field to the measure pill in the rows and finally, double click title and click insert and insert the parameter. 

## Edit and delete parameters. 

To edit the parameter: Go to the parameter pill, right click, edit, and then change the range of values. 

To delete the parameter: Note it would break controls based on the parameter. We go to the parameter, right click, delete.

## Define a group

For example, we would like to group two rows together. We would click the two observations, use the super tool box that appears, and then click group members. It will group the string using & in between the two field names.

## Define a group for others value. 

The first way: 
Click the field that we're group, click edit group. Click include other in the lower left corner. 

The second way:
Click an existing pill in the rows or column area and then click "edit pill" and click 'include other' .

## Finding members of a group. 

Go to the field that is grouping its values. Click on edit group and then click "find" to see all the members. 

## Edit or delete a group. 

CLick on the pill with group in question, right click, edit group, and then we would click the field we wanat to add to a group and click "add to": `group_name`. If we want to remove a member, select and click ungroup. If we want to change alias of a group, click rename, change name, go ahead.

If we want to change name of alias from the data vis then just double click the name. If we want to ungroup everything, click on the group, click the ungroup button in the taskbar above.

## Dashboards

* After creating a dashboard, we can resize by just simply resizing....drag and shit. 
* To add a dashboard title, look at the dashboard pane => Click show dashboard title. 
* We can also adjust formatting for different viewing devices: Go to the dashboard pane => Click the phone item. Can press device preview to preview it. 
* TO replace worksheet quickly: Click the sheet in the dashboard => Go to dashboard pane and hover over the replacing sheet and see "Swap panes"
* To define a filter action: Say we want to filter based on a table with certain criteria... => Add worksheet to dashboard => Right click and make it a float => Look on the black bar on the right and click "use as filter". 
* Highlight action: click worksheet on dashboard => Go to dashboard => Actions => Add action => add highlight. This will allow for the dashboard interaction to allow you to click a field in a worksheet and return a highlighted value for all other interacting sheets. 
* Parameter action: Go to data panel and click arrow on top right => click create parameter => Go to Dashboard => Click actions => Change parameter => define target parameter => Connect it to a field. We can now have a reference line with the value linked to the parameter. Now, if we click a point on a point on a graph, it will create a parameter/reference line through the point.
